```python
import pandas as pd
import pandas as pd 
from sklearn.preprocessing import LabelEncoder 

initial_file_path = "data/loan_data_original.csv"
df_to_change = pd.read_csv(initial_file_path)
df_to_change.info()
```

```python
df_to_change.head()
```

Comment:

* categorical features: [person_gender, person_education, person_home_ownership, loan_intent, previous_loan_defaults_on_file]

```python
# mapping person_gender

mapping_gender = {"male": 0, "female": 1}
df_to_change["person_gender"] = df_to_change["person_gender"].map(mapping_gender)
df_to_change.head()
```

```python
# mapping person_education
df_to_change["person_education"].unique()

mapping_education = {"Master": 0, "High School": 1, "Bachelor": 2, "Associate": 3, "Doctorate": 4}
df_to_change["person_education"] = df_to_change["person_education"].map(mapping_education)
df_to_change.head()
```

```python
# mapping person_home_ownership
df_to_change["person_home_ownership"].unique()

mapping_home_ownership = {"RENT": 0, "OWN": 1, "MORTGAGE": 2, "OTHER": 3}
df_to_change["person_home_ownership"] = df_to_change["person_home_ownership"].map(mapping_home_ownership)
df_to_change.head()
```

```python
# mapping loan_intent
df_to_change["loan_intent"].unique()

mapping_loan_intent = {"PERSONAL": 0, "EDUCATION": 1, "MEDICAL": 2, "VENTURE": 3, "HOMEIMPROVEMENT": 4, "DEBTCONSOLIDATION": 5}
df_to_change["loan_intent"] = df_to_change["loan_intent"].map(mapping_loan_intent)
df_to_change.head()
```

```python
# mapping previous_loan_defaults_on_file
df_to_change["previous_loan_defaults_on_file"].unique()

mapping_previous_loan = {"No": 0, "Yes": 1}
df_to_change["previous_loan_defaults_on_file"] = df_to_change["previous_loan_defaults_on_file"].map(mapping_previous_loan)
df_to_change.head()
```

Now lets save df to .csv

```python
df_to_change.to_csv("data/loan_data_all_numerical.csv", index=False)
```
